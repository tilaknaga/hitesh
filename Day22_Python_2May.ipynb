```python
def str1(x):
    a = {"count_upper":0, "count_lower":0}
    for c in x:
        if c.isupper():
            a["count_upper"]+= 1
        elif c.islower():
            a["count_lower"]+= 1
        else:
            pass
    print("The original string:", x)
    print("The number of upper characters:", a["count_upper"])
    print("The number of Lower characters:", a["count_lower"])
str1(str(input("Enter the string:")))
```
